```python
import os
import pickle

# Path to the folder containing models
model_folder = "D:\Finance_saving_project\model\Model"

# Dictionary to store loaded models
models = {}

# Check if the folder exists
if os.path.isdir(model_folder):
    # Iterate over all files in the folder
    for file_name in os.listdir(model_folder):
        # Check if the file is a pickle file
        if file_name.endswith('.pkl'):
            model_path = os.path.join(model_folder, file_name)
            try:
                # Load the model
                with open(model_path, 'rb') as file:
                    model_name = file_name.split('.')[0]  # Use file name (without extension) as key
                    models[model_name] = pickle.load(file)
                    print(f"Loaded model: {model_name} from {file_name}")
            except Exception as e:
                print(f"Error loading model {file_name}: {e}")
else:
    print(f"Model folder '{model_folder}' does not exist.")

# Access loaded models dynamically
print("\nLoaded Models:")
for name, model in models.items():
    print(f"Model Name: {name}, Model Object: {model}")
```
